```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.barplot(x='sex', y='total_bill', data=tips) # by default estimator is mean of y
```

```python
import numpy as np
```

```python
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
```

#### Count Plot

```python
sns.countplot(x='sex', data=tips)
```

#### Box Plot

```python
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='sex')
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker', palette="Set3")
```

#### Violin Plot

```python
sns.violinplot(x='day', y='total_bill', data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

#### Strip Plot

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True, hue='sex', dodge=True)
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips, hue='sex')
```

#### Factorplot

```python
sns.factorplot(x='day', y='total_bill', data=tips, kind='bar')
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='bar')
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='boxen')
```

```python
sns.catplot(x='day', data=tips, kind='count')
```
