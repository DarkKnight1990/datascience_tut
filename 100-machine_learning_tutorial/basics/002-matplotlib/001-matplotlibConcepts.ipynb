```python
import matplotlib.pyplot as plt
```

```python
## To see the plots inside the notebook
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
# FUNCTIONAL approach
plt.plot(x, y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
plt.show()
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')
```

```python
# Object Oriented Method
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x, y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.set_title('LARGE PLOT')
axes2.set_title('SMALL PLOT')

axes1.plot(x,y)
axes2.plot(y,x)
```

```python
# creating subplots
fig, axes = plt.subplots(nrows=1, ncols=2)

# for curr_ax in axes:
#     curr_ax.plot(x,y)
# axes.plot(x, y)

axes[0].set_title('First Plot')
axes[0].plot(x, y)

axes[1].set_title('Second Plot')
axes[1].plot(y, x)
plt.tight_layout()
```

**Figure size, Aspect Ratio and DPI**

```python
fig = plt.figure(figsize=(8,2))

axes = fig.add_axes([0,0,1, 1])
axes.plot(x, y)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig.savefig('my_testpic.svg', dpi=200)
```

```python
fig = plt.figure()
ax = fig.add_axes([0.1,0.2,0.8,0.8])
ax.set_title('LEGENDS Example')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.plot(x, x**2, label='X Squared')
ax.plot(x, x**3, label='X Cubed')
ax.legend(loc=0)
```

### Plot Appearance

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
# ax.plot(x,y, color='orange')
ax.plot(x,y, color='purple', linewidth=15, alpha=0.5) #RGB Hex Code
ax.plot(x,x**3, color='purple', linewidth=15)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**3, color='purple', lw=3, linestyle='-', marker='s')
```
